```python
import torch
import matplotlib.pyplot as plt

# VAEAC checkpoints are usually torch files, not pickle
results_path = "/content/vaeac/celeba_model/best_checkpoint.tar"  # Adjust path

data = torch.load(results_path, map_location="cpu")

val_iwae = data["validation_iwae"]
train_vlb = data["train_vlb"]

plt.figure(figsize=(10,4))
plt.plot(val_iwae, label="Validation IWAE", marker="o")
plt.plot(train_vlb, label="Train VLB", marker="x")
plt.xlabel("Validation checkpoint")
plt.ylabel("Loss")
plt.title("Validation IWAE vs Train VLB")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pickle
import matplotlib.pyplot as plt

# Map each run name to its .pkl result file
runs = {
    "Learned Alpha": "/content/vaeac/results/alpha_learned/iwae_and_vlb.pkl",
    "Alpha = 0.0":   "/content/vaeac/results/alpha_0.0/iwae_and_vlb.pkl",
    "Alpha = 0.5":   "/content/vaeac/results/alpha_0.5/iwae_and_vlb.pkl",
    "Alpha = 1.0":   "/content/vaeac/results/alpha_1.0/iwae_and_vlb.pkl",
}

plt.figure(figsize=(8,4))

for label, path in runs.items():
    try:
        with open(path, "rb") as f:
            data = pickle.load(f)
        alpha_history = data["alpha_history"]
        plt.plot(alpha_history, marker="*", label=label)
    except Exception as e:
        print(f"Could not load {path}: {e}")

plt.xlabel("Validation checkpoint")
plt.ylabel("KL Alpha")
plt.title("KL Alpha Progression Across Runs")
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig("alpha_progression.png", dpi=200)
```

```python
import torch

# Path to your uploaded checkpoint file
ckpt_path = "/content/vaeac/celeba_model/best_checkpoint.tar"
ckpt = torch.load(ckpt_path, map_location="cpu")

# Keys inside the checkpoint
print("Checkpoint keys:", list(ckpt.keys()))

# 1. Epoch
epoch = ckpt.get("epoch", None)
print(f"\nSaved at epoch: {epoch}")

# 2. Training history
train_vlb = ckpt.get("train_vlb", None)
validation_iwae = ckpt.get("validation_iwae", None)

if train_vlb is not None and validation_iwae is not None:
    print(f"\nNumber of train VLB records: {len(train_vlb)}")
    print(f"Number of validation IWAE records: {len(validation_iwae)}")

    print("\nLast 5 Train VLBs:", train_vlb[-5:])
    print("Last 5 Validation IWAEs:", validation_iwae[-5:])
else:
    print("No training/validation history found.")

# 3. Model state
print("\nModel state dict keys:", list(ckpt["model_state_dict"].keys())[:5], "...")

# 4. Optimizer state
print("Optimizer state dict keys:", list(ckpt["optimizer_state_dict"].keys())[:5], "...")

# If you want a plot of metrics, uncomment and run this in a notebook:
import matplotlib.pyplot as plt
# plt.plot(train_vlb, label='Train VLB')
# plt.plot(validation_iwae, label='Validation IWAE')
plt.legend(); plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import os

# List your inpainting result directories (one per alpha)
alpha_dirs = {
    "Alpha 0.0": "celeba_inpaintings_alpha_0.0",
    "Alpha 0.5": "celeba_inpaintings_alpha_0.5",
    "Alpha 1.0": "celeba_inpaintings_alpha_1.0",
    "Alpha learned": "celeba_inpaintings_alpha_learned",
}

n_show = 3       # how many examples to display
sample_num = 0   # which inpainted sample to show (0, 1, 2, ...)

for i in range(n_show):
    fig, axs = plt.subplots(1, 2 + len(alpha_dirs), figsize=(17, 4))

    # Use the first directory just for groundtruth and masked input
    ref_dir = list(alpha_dirs.values())[0]
    files = [
        (os.path.join(ref_dir, f"{i:05d}_groundtruth.jpg"), "Groundtruth"),
        (os.path.join(ref_dir, f"{i:05d}_input.jpg"), "Masked input")
    ]

    # Add sample_num-th inpainted image for each alpha
    for label, d in alpha_dirs.items():
        sample_path = os.path.join(d, f"{i:05d}_sample_{sample_num:03d}.jpg")
        files.append((sample_path, f"{label}"))

    for j, (fname, title) in enumerate(files):
        try:
            img = Image.open(fname)
            axs[j].imshow(img)
            axs[j].set_title(title)
            axs[j].axis("off")
        except Exception as e:
            axs[j].set_title(f"{title}\n(Not found)")
            axs[j].axis("off")
    plt.tight_layout()
    plt.show()
```
